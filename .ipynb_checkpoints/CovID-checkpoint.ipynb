The data is hosted by John Hopkins University and can be obtained through `!pip install covid-data-api`.

```python
from utils import *
# DataFrame with cases for all available countries
df = {}
for loc in api.show_available_countries():
    df[loc] = get_df(loc)
    df[loc] = add_cases(df[loc]) # calculate 'New cases' and 'Active cases' and add to the DataFrame
```

# 1. China

```python
locations = ['China']
for loc in locations:
    plt_cases(df, loc)
```

# 2. United States

```python
locations = ['US']
for loc in locations:
    plt_cases(df, loc)
```

# 3. Europe

```python
locations = ['Italy', 'Spain', 'Germany', 'France']
for loc in locations:
    plt_cases(df, loc)
```

# 4. Russia, Ukraine, Belarus

```python
locations = ['Russia', 'Ukraine', 'Belarus']
for loc in locations:
    plt_cases(df, loc)
```

# 5. Some other locations

```python
locations = ['Iran', 'United Kingdom', 'Turkey', 'Switzerland','Belgium', 'Netherlands', 'Canada',
             'Austria', 'Korea, South', 'Portugal', 'Brazil', 'Israel', 'Sweden', 'Australia', 'Norway']
for loc in locations:
    plt_cases(df, loc)
```
